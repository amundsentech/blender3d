```python
## grab start location information
collars_collar_col='Hole ID'
x_col='Easting' 
y_col='Northing'
z_col='Elevation'


## grab survey columns will be user choice
survey_collar_col='Hole ID'
dip_col='Dip'
azi_col='Azimuth'
survey_depth_col='Depth_m'

### grab the data depth
data_collar_col='Hole ID'
data_z_col1='From_m'
data_z_col2='To_m'
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
```

```python
def calc_xyz(azi,dip,depth):
    # scalar length be is the distance from the previous point or depth in geo terms

    x_hat = depth*np.cos(dip)*np.sin(azi)
    y_hat = depth*np.cos(dip)*np.cos(azi) 
    z_hat = depth*np.sin(dip) 
    return x_hat,y_hat,z_hat
```

```python
import matplotlib as mpl
from matplotlib import cm
class MplColorHelper:

  def __init__(self, cmap_name, start_val, stop_val):
    self.cmap_name = cmap_name
    self.cmap = plt.get_cmap(cmap_name)
    self.norm = mpl.colors.Normalize(vmin=start_val, vmax=stop_val)
    self.scalarMap = cm.ScalarMappable(norm=self.norm, cmap=self.cmap)


  def get_rgb(self, val):
    return self.scalarMap.to_rgba(val)
```

```python
cmap=MplColorHelper(cmap_name='coolwarm',start_val=0,stop_val=1)
```

```python
drill_data=[]
for name,group in lith.groupby('Hole ID'):
    surv=survey[survey[survey_collar_col]==name]
    start_data=collar[collar[collars_collar_col]==name]
    for i in surv[survey_depth_col].values:
        group.loc[-1,data_z_col1]=i
        group=group.reset_index(drop=True)

    group=group.sort_values(data_z_col1)
    data=pd.merge(group,surv,left_on=[data_z_col1],right_on=[survey_depth_col],how = 'left',
                                                                                suffixes=('', '_delme')  # Left gets no suffix, right gets something identifiable
                                                                                )
    data = data[[c for c in data.columns if not c.endswith('_delme')]]
    data[[azi_col,dip_col]]=data[[azi_col,dip_col]].fillna(method='ffill').fillna(method='bfill')
    drop=data[data[data_collar_col].isna()].index
    data=data.drop(drop).dropna(axis=1,how='any')
    
    ## convert angles to radians
    data [azi_col]=data [azi_col].apply(radians)
    data [dip_col]=data [dip_col].apply(radians)

    ## grab starting coordinates
    start_x=start_data[x_col].values[0]
    start_y=start_data[y_col].values[0]
    start_z=start_data[z_col].values[0]
    print('Starting xyz',start_x,start_y,start_z)

    # grab starting depth
    last_depth=data.loc[0,data_z_col1]

    data.loc[0,x_col]=start_x
    data.loc[0,y_col]=start_y
    data.loc[0,z_col]=start_z

    ax.scatter(start_x,start_y,start_z)
    
    for i, row in data.iterrows():
        if i>0:
            new_depth=row[data_z_col1]

            azi=row[azi_col]
            dip=row[dip_col]

            distance=np.abs(new_depth-last_depth)

        
            x_hat,y_hat,z_hat=calc_xyz(azi=azi,dip=dip,depth=distance)


            x=start_x+x_hat
            y=start_y+y_hat
            z=start_z+z_hat

            data.loc[i,x_col]=x
            data.loc[i,y_col]=y
            data.loc[i,z_col]=z
            
            start_x=x
            start_y=y
            start_z=z

            last_depth=new_depth

    ax.plot(data[x_col],data[y_col],data[z_col])
    display(data)
    drill_data.append(data)
drill_holes=pd.concat(drill_data)
drill_holes

    
    
```
