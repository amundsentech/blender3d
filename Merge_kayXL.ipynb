```python
import pandas as pd
import glob

from tqdm.autonotebook import tqdm
```

```python
path='/Users/sebastiancoombs/Library/CloudStorage/GoogleDrive-sebastianbcoombs@gmail.com/Shared drives/AMC Projects/_AZ_Kay/_Drilling/_Logs/**xlsx'
```

```python
xl_files=glob.glob(path)
```

```python
sheet_names=pd.ExcelFile(xl_files[0]).sheet_names
```

```python
file=pd.read_excel(xl_files[0],sheet_name=sheet_names[0])
file.set_index(file.columns[0]).T
```

```python
kay_collars=[]
for xl_file in tqdm(xl_files):
    try:
        sheet_names=pd.ExcelFile(xl_file).sheet_names
        collar_sheet=[c for c in sheet_names if 'collar' in c.lower()][0]
        file=pd.read_excel(xl_file,sheet_name=collar_sheet)
        file=file.set_index(file.columns[0]).T
        hole_id=[c for c in file.columns if 'hole' in c.lower()][0]
        file=file.set_index(hole_id)
        # display(file)
        kay_collars.append(file)
    except: pass
```

```python

collars=pd.concat(kay_collars,axis=0)
collars=collars[collars.index.isna()==False]
collars=collars[collars['Collar UTM E'].isna()==False]
collars.shape
```
